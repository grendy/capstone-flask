```python
from flask import Flask, render_template
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from io import BytesIO
import base64
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
playstore = pd.read_csv('data/googleplaystore.csv')
```

```python
# Hapus data yang duplikat berdasarkan kolom App, dengan tetap keep data pertama (hint : gunakan parameter subset)
playstore.drop_duplicates(subset = 'App', keep = 'first', inplace=True) 
```

```python
# bagian ini untuk menghapus row 10472 karena nilai data tersebut tidak tersimpan pada kolom yang benar
playstore.drop([10472], inplace=True)
```

```python
# Cek tipe data kolom Category. Jika masih tersimpan dengan format tipe data yang salah, ubah ke tipe data yang sesuai
playstore.Category = playstore['Category'].astype('category')

playstore
```

```python
# Pada kolom Installs Buang tanda koma(,) dan tanda tambah(+) kemudian ubah tipe data menjadi integer
playstore.Installs = playstore['Installs'].apply(lambda x: x.replace(',',''))
playstore.Installs = playstore['Installs'].apply(lambda x: x.replace('+',''))
```

```python
playstore.Installs
```

```python
# bagian untuk mengubah tipe data Installs
playstore.Installs = pd.to_numeric(playstore.Installs)
```

```python
playstore.Installs
```

```python
# Bagian ini untuk merapikan kolom Size, Anda tidak perlu mengubah apapun di bagian ini
playstore['Size'].replace('Varies with device', np.nan, inplace = True ) 
playstore.Size = (playstore.Size.replace(r'[kM]+$', '', regex=True).astype(float) * \
             playstore.Size.str.extract(r'[\d\.]+([kM]+)', expand=False)
            .fillna(1)
            .replace(['k','M'], [10**3, 10**6]).astype(int))
playstore['Size'].fillna(playstore.groupby('Category')['Size'].transform('mean'),inplace = True)
```

```python
# Pada kolom Price, buang karakater $ pada nilai Price lalu ubah tipe datanya menjadi float
playstore.Price = playstore['Price'].apply(lambda x: x.replace('$',''))
playstore.Price = playstore['Price'].astype(float)
```

```python
playstore.Price
```

```python
playstore[['Reviews','Size','Installs']] = \
playstore[['Reviews','Size','Installs']].astype('int')
```

```python
cat_order = df2.groupby('Category').agg({'App': 'count'}).reset_index().rename({'App':'Total'}, axis=1).sort_values('Total',ascending=False).head()

X = cat_order.Category
Y = cat_order.Total
my_colors = 'rgbkymc'

fig = plt.figure(figsize=(8,3),dpi=300)
fig.add_subplot()

plt.barh(X,Y, color=my_colors)

plt.savefig('cat_order.png',bbox_inches="tight")
```
